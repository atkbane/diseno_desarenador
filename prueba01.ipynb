# CALCULANDO EL DESARENADOR

```python
# CREANDO UNA FUNCION PARA EL CALCULO DE DIMENSIONES!

def diseno_desarenador_003(caudal, dia_part, alto=False, ancho=False, rel=1.5, vel_sudry=3.5, n_naves=3):
    '''
    funcion para calcular las dimensiones de un desarenador
    caudal: caudal de diseño del proyecto
    dia_part = diametro de la particula a desarenar mm
    vel_sudry = velocidad de caida de particula segun sudry en cm/seg
    '''
    # DATOS INICIALES
    caudal_1 = round_decimals_up(caudal / n_naves, 2) # caudal de diseño por nave
    d_particula = dia_part # milimetros
    velo_flujo = vel_CAMP(d_particula) # velocidad del flujo en cm/seg recomendada obtenida unando formula de camp
    velo_caida_sudry = vel_sudry / 100 # velocidad de caida de la particula en cm/seg
    relacion = rel # relacion entre h (alto) y b (base)

    # empezamos con los calculos para la cantidad de naves inicial
    
    area_1 = caudal_1 / (velo_flujo / 100)

    if alto and ancho:
        alto_1 = alto
        base_1 = ancho

    elif alto:
        alto_1 = alto
        base_1 = round_decimals_up(area_1 / alto_1, 1)
    
    elif ancho:
        base_1 = ancho
        alto_1 = round_decimals_up(area_1 / base_1, 1)

    else:
        alto_1 = round_decimals_up((area_1 / relacion) ** 0.5, 1)
        base_1 = round_decimals_up(alto_1 * relacion)

    velo_flujo_1 = round_decimals_up(caudal_1 / (alto_1 * base_1), 2) # velocidad con la cantidad de naves inicial

    long_desar_1 = l_sedim(alto_1, velo_flujo_1, velo_caida_sudry)

    print(f"""las medidas del desarenador de {n_naves} son las siguientes
    - Ancho: {base_1} m
    - Alto: {alto_1} m
    - Velocidad: {velo_flujo_1} m/seg
    - Longitud: {long_desar_1} m
    """)

    # ahora reducimos la cantidad de naves en 1

    # DATOS INICIALES
    n_naves_2 = n_naves - 1 # cantidad de naves activas
    caudal_2 = round_decimals_up(caudal / n_naves_2, 2)

    # nueva area de flujo para el nuevo caudal
    area_2 = caudal_2 / (velo_flujo / 100)

    # dimenisones del flujo
    base_2 = base_1
    alto_2 = round_decimals_up(area_2 / base_2, 1)

    # velocidad del flujo con las dimensiones finales
    velo_flujo_real_2 = round_decimals_up(caudal_2 / (base_2 * alto_2), 2)

    # calculo de la longitud del desarenador con dos naves
    long_desar_2 = l_sedim(alto_2, velo_flujo_real_2, velo_caida_sudry)

    print(f"""las medidas del desarenador de {n_naves} naves (1 en limpieza) son las siguientes:
    - Ancho: {base_2} m
    - Alto: {alto_2} m
    - Velocidad: {velo_flujo_real_2} m/seg
    - Longitud : {long_desar_2} m""")

diseno_desarenador_003(caudal=9.5, dia_part=0.3, ancho=5)
```
